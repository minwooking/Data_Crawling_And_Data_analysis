```python
tokens = []
for words in tqdm(df.내용정제):
    word = okt.nouns(str(words))
    tokens.append(word)
```

```python
tokenss = []
for a in tqdm(tokens):
    for b in a:
        tokenss.append(b)
```

```python
new_tokens =[]
for token in tokenss:
    new_token = re.sub('[^가-힣 ]', '', token)
    if new_token:                       # ''는 포함시키지 말 것
        new_tokens.append(new_token)
```
