```python
urllist = ['https://search.shopping.naver.com/catalog/27377160522?query=%EC%B2%AD%EB%A3%A1%EB%86%8D%EC%9B%90&NaPm=ct%3Dl2egmi6g%7Cci%3Daac2c02ea8771b0f19140701faa7c430562ee625%7Ctr%3Dslsl%7Csn%3D95694%7Chk%3Daabc9919286ce2cdf8cb1beff7fbd3f5c8bba2a6']
평점 = []
내용 = []
작성시간 = []
# 상품명 = []

for k,urls in enumerate(urllist):
    driver.get(urls)
    time.sleep(1)
    try:
        for j in tqdm(range(14)):
            if j==0:
                for i in tqdm(range(1,12)):
                    driver.find_element_by_xpath(f'//*[@id="section_review"]/div[3]/a[{i}]').click()
                    html = driver.page_source
                    soup = BeautifulSoup(html, 'html.parser')
                    content = soup.select_one('#section_review > ul')
                    time.sleep(1)
                    for p in range(len(content.find_all('li'))):
                        평점.append(content.find_all('li')[p].text[:6])
                        내용.append(content.find_all('li')[p].text[6:])
                        작성시간.append(re.sub(r'[^0-9]','',content.find_all('li')[p].text[6:]))
                time.sleep(1)
            else:
                for a in tqdm(range(2,13)):
                    driver.find_element_by_xpath(f'/html/body/div/div/div[2]/div[2]/div[2]/div[3]/div[5]/div[3]/a[{a}]').click()
                    html = driver.page_source
                    soup = BeautifulSoup(html, 'html.parser')
                    content = soup.select_one('#section_review > ul')
                    time.sleep(1)
                    for k in range(len(content.find_all('li'))):
                        평점.append(content.find_all('li')[k].text[:6])
                        내용.append(content.find_all('li')[k].text[6:])
                        작성시간.append(re.sub(r'[^0-9]','',content.find_all('li')[k].text[6:]))
                        # if k == 0 : 상품명.append('GNM자연의품격')
                        # if k == 1 : 상품명.append('청룡농원')
                time.sleep(1)
    except:
        pass    
```
