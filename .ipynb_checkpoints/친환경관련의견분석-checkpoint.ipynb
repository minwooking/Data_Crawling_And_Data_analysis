```python
df = pd.read_csv('미주미.csv')
```

```python
df['내용'].value_counts()
```

```python
result = ''
for i,content in tqdm(enumerate(df['내용'])):
    try:
        result+=str(content.split('\n'))
    except: 
        result+=str(content)
    
```

```python
results = [ for i in if result[i] =! '']
```

```python
result[:10]
```

```python
okt.nouns(result)[:10]
```

```python
token = okt.nouns(result)
```

```python
# 한글 이외의 글자는 제거
import re
new_tokens =[]
for toke in token:
    new_token = re.sub('[^가-힣 ]', '', toke)
    if new_token:                       # ''는 포함시키지 말 것
        new_tokens.append(new_token)
```

```python
cafe_text = nltk.Text(new_tokens, name='친환경')
plt.figure(figsize=(15,6))
cafe_text.plot(50)
plt.show()
```

```python
stop_words = '것 투자 생각 수 이 주 더 때 등 말 그 글 전 의 제 저 위 중 요 점 및 를 내 분 개 로 미국 친환경 주식 기업 종목 회사 주가 지금 정도'
stop_list = stop_words.split()
```

```python
new_tokens = [token for token in new_tokens if token not in stop_list]
```

```python
cafe_text2 = nltk.Text(new_tokens, name='친환경')
plt.figure(figsize=(15,6))
cafe_text2.plot(50)
plt.show()
```

```python
data = cafe_text2.vocab().most_common(300)
data[:10]
```

```python
wc = WordCloud(
    font_path='NanumGothic.ttf',
    relative_scaling=0.2,
    background_color='white'
).generate_from_frequencies(dict(data))
```

```python
plt.figure(figsize=(18,12))
plt.imshow(wc)
plt.axis('off')
plt.show()
plt.savefig('전기차시장분석')
```
